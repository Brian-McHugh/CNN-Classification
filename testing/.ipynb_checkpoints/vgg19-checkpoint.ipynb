```python
# Dependencies
import matplotlib.pyplot as plt
%matplotlib inline

import os
import numpy as np
import tensorflow as tf
from PIL import Image

import keras
from keras.preprocessing import image
from keras.applications.vgg19 import (
    VGG19, 
    preprocess_input, 
    decode_predictions
)
```

```python
# Load the VGG19 model
# https://keras.io/applications/#VGG19
model = VGG19(include_top=True, weights='imagenet')
```

```python
# Define default image size for VGG19
image_size = (224, 224)
```

```python
# Load the image and resize to default image size
image_path = os.path.join("..", "images", "chair_on_back.jpg")
img = image.load_img(image_path, target_size=image_size)
plt.imshow(img)
```

```python
# Preprocess image for model prediction
# This step handles scaling and normalization for VGG19
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
x.shape
```

```python
# Make predictions
predictions = model.predict(x)
print('Predicted:', decode_predictions(predictions, top=5))
plt.imshow(img)
```

```python
# Function to rotate images so they look upright
def rotate_img(img_path, rt_degr):
    chair = Image.open(img_path)
    return chair.rotate(rt_degr, expand=1)
```

```python
# Create the new rotated image
img_rt_270 = rotate_img("../images/chair_on_back.jpg", 270)

# Save the rotated image to our images folder (uncomment to resave)
img_rt_270.save("../images/brown_chair_upright.jpg")
```

```python
# Run prediction again on our new, rotated image
# Load the image and resize to default image size
image_path = os.path.join("..", "images", "brown_chair_upright.png")
img = image.load_img(image_path, target_size=image_size)
```

```python
# Refactor above steps into reusable function
def predict(image_path):
    """Use VGG19 to label image"""
    img = image.load_img(image_path, target_size=image_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    predictions = model.predict(x)
    plt.imshow(img)
    print('Predicted:', decode_predictions(predictions, top=5))
```

```python
# See if results change for this new version of the brown chair
image_path = os.path.join("..", "images", "brown_chair_upright.png")
predict(image_path)
```

```python
# Try process again with a chair lying on its side instead of its back
# Using the same PIL rotation function
image_path = os.path.join("..", "images", "chair_on_left_side.jpg")
predict(image_path)
```
